```python
import cobra
```

```python
import escher
```

```python
from cobra import *
```

```python
from escher import Builder
```

```python
model = Model('Test')
```

```python
GLC = Metabolite('GLC', compartment='c')
AA = Metabolite('AA', compartment='c')
Biomass = Metabolite('Biomass', compartment='c')
```

```python
##### r0 : ======> GLC
r0 = Reaction('r0')
r0.name = 'R0'
r0.lower_bound = 1
r0.upper_bound = 1
```

```python
##### r1 : GLC ======> AA
r1 = Reaction('r1')
r1.name = 'R1'
r1.lower_bound = 0
r1.upper_bound = 1000
```

```python
##### r2 : AA ======> Biomass
r2 = Reaction('r2')
r2.name = 'R2'
r2.lower_bound = 0
r2.upper_bound = 1000
```

```python
##### r3 : Biomass ======>
r3 = Reaction('r3')
r3.name = 'R3'
r3.lower_bound = 0
r3.upper_bound = 1000
```

```python
r0.add_metabolites({GLC: 1})
```

```python
r1.add_metabolites({GLC: -1, AA: 1})
```

```python
r2.add_metabolites({AA: -9.09, Biomass: 1})
```

```python
r3.add_metabolites({Biomass: -1})
```

```python
model.add_reactions([r0, r1, r2, r3])
```

```python
model.objective = 'r3'
```

```python
model.optimize()
```

```python
model.summary()
```

```python
cobra.io.save_json_model(model, "task2.json")
```
